Упражнение 4

```python
fig, axes = plt.subplots(2, 3, figsize=(12, 8))

# Задаем промежуток [−5;5]
x = np.linspace(-5, 5, 400)
y1 = np.abs(np.abs(x) - 2)

# Функции для остальных графиков
# график получается сдвигом исходного на 2 единицы вправо
def f1(x):
    return np.abs(np.abs((x - 2)) - 2)

# график получается сдвигом исходного на 2 единицы влево
def f2(x):
    return np.abs(np.abs((x + 2)) - 2)

# график получается растяжением исходного вдоль оси х в два раза
def f3(x):
    return np.abs(np.abs(2 * x) - 2)

# график получается сжвтием исходного вдоль оси х в два раза
def f4(x):
    return np.abs(np.abs(0.5 * x) - 2)

# график идентичен исходному
def f5(x):
    return np.abs(np.abs(-x) - 2)

# Строим графики на подобластях
axes[0, 0].plot(x, y1, label='y = f(x) - 2', color='blue')
axes[0, 1].plot(x, f1(x), label='y = f(x - 2) - 2', color='green')
axes[0, 2].plot(x, f2(x), label='y = f(x + 2) - 2', color='red')
axes[1, 0].plot(x, f3(x), label='y = 2f(x)', color='purple')
axes[1, 1].plot(x, f4(x), label='y = 0.5f(x)', color='orange')
axes[1, 2].plot(x, f5(x), label='y = f(-x)', color='brown')

# Добавляем заголовки и легенду
axes[0, 0].set_title('y = f(x) - 2')
axes[0, 1].set_title('y = f(x - 2) - 2')
axes[0, 2].set_title('y = f(x + 2) - 2')
axes[1, 0].set_title('y = f(2x) - 2')
axes[1, 1].set_title('y = f(0.5x) - 2')
axes[1, 2].set_title('y = f(-x) - 2')

for ax in axes.flatten():
    ax.grid(True)
    ax.legend()

# Настройка размеров и расположения подобластей
plt.tight_layout()

plt.show()
```

Упражнение 5

```python
import mpmath

# Установим относительную погрешность
mpmath.mp.dps = 4  # 4 значащих цифры (приближенно 0.1% точности)

x = np.linspace(-2, 2, 400)

shx = [mpmath.sinh(xi) for xi in x]
chx = [mpmath.cosh(xi) for xi in x]

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

ax1.plot(x, shx, label='sh(x)', color='blue')
ax1.axhline(0, color='black', linewidth=0.5)
ax1.axvline(0, color='black', linewidth=0.5)
ax1.grid(True, linestyle='--', alpha=0.7)
ax1.set_title('График функции sh(x)')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.legend()

# Настраиваем оси и сетку для второй подобласти
ax2.plot(x, chx, label='ch(x)', color='red')
ax2.axhline(0, color='black', linewidth=0.5)
ax2.axvline(0, color='black', linewidth=0.5)
ax2.grid(True, linestyle='--', alpha=0.7)
ax2.set_title('График функции ch(x)')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.legend()

plt.tight_layout()

plt.show()
```

```python
x = np.linspace(0, np.pi, 400)
y1 = np.cos(x)
y2 = np.arccos(x)
x_ = np.arange(0, 2, 0.5)
y_ = x_

fig, ax = plt.subplots(figsize=(8, 6))

ax.plot(x, y1, linestyle='--', color='green', label='y = cos(x)')
ax.plot(x, y2, linestyle='-.', color='red', label='y = arccos(x)')
plt.plot(x_, y_, color='blue', linewidth=1, label='Симметричная прямая')

ax.axhline(0, color='black', linewidth=1)
ax.axvline(0, color='black', linewidth=1)
ax.grid(True, linestyle='--', alpha=0.7)
ax.axis('equal')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Графики функции cos(x) и обратной функции arccos(x) с симметричной прямой')
ax.legend()
plt.show()
```

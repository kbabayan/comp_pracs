Упражненние 4

```python
import matplotlib.pyplot as plt

z1=np.array([1+3j])
z2=np.array([3+4j])
z3=np.array([z1+z2])
z4=np.array([z1-z2])
orig = np.zeros(1)

plt.quiver(orig,orig, np.real(z1), np.imag(z1),angles='xy', scale_units='xy', scale=1, color="green")
plt.quiver(orig,orig, np.real(z2), np.imag(z2),angles='xy', scale_units='xy', scale=1, color="black")
plt.quiver(orig,orig, np.real(z3), np.imag(z3),angles='xy', scale_units='xy', scale=1, color="blue")
plt.quiver(orig,orig, np.real(z4), np.imag(z4),angles='xy', scale_units='xy', scale=1, color="red")

plt.axis([-8,8,-8,8])
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')

plt.grid(True)

plt.title('Complex numbers')
plt.xlabel('Re(z)')
plt.ylabel('Im(z)')
plt.show()
```

Упражнение 5

```python
z = 2*np.sqrt(3)-2j
r = np.abs(z)
phi=np.angle(z)
k = np.arange(0,4)
zroot = r**(1/4)*( np.cos((phi+2.0*np.pi*k)/4) +1j*np.sin((phi+2.0*np.pi*k)/4))
orig = np.zeros(1)

plt.title('$\sqrt[4]{2\sqrt{3}-2j}$ complex roots')
plt.xlabel('Re(z)')
plt.ylabel('Im(z)')
plt.grid(True)
plt.plot(np.real(zroot), np.imag(zroot), '*g') 
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')

plt.axis('equal')
t = np.linspace(0,2*np.pi,100)
plt.plot(r**(1/4)*np.cos(t),r**(1/4)*np.sin(t),':')
```

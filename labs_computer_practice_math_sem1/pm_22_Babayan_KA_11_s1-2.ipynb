```python
import scipy.optimize as optimize

def f(x):
    return np.exp(-x) * np.sin(3 * np.pi * x)

def der(x_podst):
    y = -1.0 * np.sin(3 * np.pi * x_podst) * np.exp(-x_podst) + 3 * np.pi * np.cos(3 * np.pi * x_podst) * np.exp(-x_podst)
    return y

def checkExtremum(f, x):
    eps = 0.001
    if f(x) - f(x - eps) > 0:
        return "Максимум"
    return "Минимум"

# построение функции
x = np.arange(0, 2, 0.01)
plt.plot(x, f(x), "r", label='f')

# построение производной
x_plt = np.arange(0, 2, 0.01)
y_plt = der(x_plt)
plt.plot(x_plt, y_plt, "b:", label="f'")

# точки экстремума
x1 = optimize.brentq(lambda x: der(x), 0, 0.25)
print("Точка экстремума 1:", x1, "-", checkExtremum(f, x1))
x2 = optimize.brentq(lambda x: der(x), 0.25, 0.5)
print("Точка экстремума 2:", x2, "-", checkExtremum(f, x2))
x3 = optimize.brentq(lambda x: der(x), 0.75, 1)
print("Точка экстремума 3:", x3, "-", checkExtremum(f, x3))
x4 = optimize.brentq(lambda x: der(x), 1, 1.25)
print("Точка экстремума 4:", x4, "-", checkExtremum(f, x4))
x5 = optimize.brentq(lambda x: der(x), 1.25, 1.5)
print("Точка экстремума 5:", x5, "-", checkExtremum(f, x5))
x6 = optimize.brentq(lambda x: der(x), 1.75, 2)
print("Точка экстремума 6:", x6, "-", checkExtremum(f, x6))

# построение точек экстремума функции
plt.plot(x1, f(x1), "ro")
plt.plot(x2, f(x2), "ro")
plt.plot(x3, f(x3), "ro")
plt.plot(x4, f(x4), "ro")
plt.plot(x5, f(x5), "ro")
plt.plot(x6, f(x6), "ro")

# установить границы оси
plt.ylim(-8, 8)

# параметры координатной плоскости
plt.xlabel("x")
plt.ylabel("y", rotation=0)
plt.grid()
plt.legend()
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.show()
```

```python
x = sp.Symbol('x')
y = np.e**(1/(x**2-1))
print(y.diff())
```

```python
def f(x):
    return np.e**(1/(x**2-1))

def der(x_podst):
    if x_podst == 1 or x_podst == -1:
        return np.inf
    else:
        y = -2.0 * np.exp(1/(x_podst**2 - 1)) * x_podst / (x_podst**2 - 1)**2
        return y

def checkExtremum(f, x):
    eps = 0.001
    if f(x) - f(x - eps) > 0:
        return "Максимум"
    return "Минимум"

# построение функции
x_range = np.linspace(-5, 5, 5000)
plt.plot(x_range, f(x_range), "b")

# построение производной
x_plt = np.arange(min(x_range), max(x_range), 0.01)
y_plt = [der(x) for x in x_plt]
plt.plot(x_plt, y_plt, "r:")

# точка экстремума
result = optimize.minimize_scalar(lambda x: abs(der(x)), bounds=(-1, 1), method='bounded')
x1 = result.x
print("Точка экстремума 1:", x1, "-", checkExtremum(f, x1), "Значение:", f(x1))

# построение точки экстремума
plt.plot(x1, f(x1), "ro")

# установить границы оси
plt.ylim(-3, 10)
plt.xlim(-4, 4)

# параметры координатной плоскости
plt.xlabel("x")
plt.ylabel("y", rotation=0)
plt.grid()
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.optimize as optimize

def f(x):
    return np.exp(1/(x**2-1))

def der_2(x):
    y = 8.0 * np.exp(1/(x**2 - 1)) * x**2 / (x**2 - 1)**3 + 4.0 * np.exp(1/(x**2 - 1)) * x**2 / (x**2 - 1)**4 - 2.0 * np.exp(1/(x**2 - 1)) / (x**2 - 1)**2
    return y

# построение функции
x_range = np.linspace(-5, 5, 5000)
plt.plot(x_range, f(x_range), "b")

# построение второй производной
x_plt = np.arange(-5, -1.01, 0.01)
y_plt = [der_2(x_val) for x_val in x_plt]
plt.plot(x_plt, y_plt, "r:")

x_plt = np.arange(-0.99, 0.99, 0.01)
y_plt = [der_2(x_val) for x_val in x_plt]
plt.plot(x_plt, y_plt, "r:")

x_plt = np.arange(1.01, 5, 0.01)
y_plt = [der_2(x_val) for x_val in x_plt]
plt.plot(x_plt, y_plt, "r:")

# точки перегиба
x1 = optimize.brentq(lambda x: der_2(x), -0.99, -0.5)
print("Точка перегиба 1:", x1, "Значение:", f(x1))
x2 = optimize.brentq(lambda x: der_2(x), 0.5, 0.99)
print("Точка перегиба 2:", x2, "Значение:", f(x2))

# построение точек перегиба
plt.plot(x1, f(x1), "ro")
plt.plot(x2, f(x2), "ro")

# установить границы оси
plt.ylim(-3, 10)
plt.xlim(-4, 4)

# параметры координатной плоскости
plt.xlabel("x")
plt.ylabel("y", rotation=0)
plt.grid()
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.show()
```

```python
x = sp.symbols('x')
k = sp.limit(np.e**(1/(x**2-1)) / x, x, sp.oo)
b = sp.limit(np.e**(1/(x**2-1)) - k * x, x, sp.oo)
print("Наклонная асимптота:", k * x + b)
```

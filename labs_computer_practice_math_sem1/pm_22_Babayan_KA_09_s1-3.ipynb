```python
def f_1(x):
    return (x**4 + 2*x**2 + x + 2)/(x**3 + 2*x**2 + x + 2)

def find_inflection_points(expr, variable):
    x = sp.symbols(variable)
    second_derivative = sp.diff(expr(x), x, 2)
    # print(sp.simplify(second_derivative))
    
    # Находим точки, где вторая производная равна нулю или не существует
    critical_points = sp.solve(sp.simplify(second_derivative), x)
    
    # Фильтруем только вещественные точки
    real_inflection_points = [point.evalf() for point in critical_points] #if point.is_real]
    
    return real_inflection_points

inflection_points = find_inflection_points(f_1, 'x')

print("Точки перегиба функции:", inflection_points)
```

```python
from scipy.constants import pi

def Maxwells_distribution(x, beta=1):
    return np.vectorize(lambda x: (np.sqrt(2/pi)) * beta**(3/2) * x**2 * np.exp((-beta*x**2)/2) if x > 0 else 0)(x)
```

```python
print(Maxwells_distribution(1))
```

```python
x = np.linspace(-100, 100, 500)
y = Maxwells_distribution(x, beta=5)

plt.plot(x, y)
plt.title('Построим функцию')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```

```python
def f_5(x):
    return np.sin(x) - x**2 * np.cos(x)
```

```python
x = np.linspace(-100, 100, 200)
plt.plot(x, f_5(x), color='blue', label='myfunc')
plt.title('Построим функцию')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

```python
x = np.linspace(-10, 10, 200)
plt.plot(x, f_5(x), color='blue', label='myfunc')
plt.title('Построим функцию')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from scipy.optimize import fsolve

def f_5(x):
    return sp.sin(x) - x**2 * sp.cos(x)

def equation_to_solve(x):
    return x**2*(1 - np.tan(x/2)**2)/(np.tan(x/2)**2 + 1) + 8*x*np.tan(x/2)/(np.tan(x/2)**2 + 1) - 2*(1 - np.tan(x/2)**2)/(np.tan(x/2)**2 + 1) - 2*np.tan(x/2)/(np.tan(x/2)**2 + 1)

def find_inflection_points(expr, variable, num_points=2, initial_guesses=np.linspace(-5, 5, 100)):
    x = sp.symbols(variable)
    second_derivative = sp.diff(expr(x), x, 2)

    # Convert the symbolic expression to a numerical function
    second_derivative_func = sp.lambdify(x, second_derivative, 'numpy')

    # Use fsolve to find numerical solutions
    numerical_inflection_points = []

    for guess in initial_guesses:
        point = fsolve(second_derivative_func, guess)
        numerical_inflection_points.extend(point)

        if len(numerical_inflection_points) >= num_points:
            break

    return numerical_inflection_points[:num_points]

inflection_points = find_inflection_points(f_5, 'x', num_points=3)

print("Точки перегиба функции:", inflection_points)
```

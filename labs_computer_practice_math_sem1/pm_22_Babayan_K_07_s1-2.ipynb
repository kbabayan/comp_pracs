```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
```

```python
# а)
x_sym = sp.Symbol('x')
f = (sp.exp(4*x_sym) - 1)/(1 - sp.exp(5*x_sym))
A = sp.limit(f, x_sym, sp.oo)
print('lim = ',A)
```

```python
x = np.linspace(0,10)
f_np = sp.lambdify(x_sym, f, 'numpy')
y_vals = f_np(x)

plt.plot(x, y_vals, color='black')
plt.title('График функции (e^(4x)-1)/(1-e^(5x))')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()
```

```python
# б)
f = ((-x_sym**2 - 3*x_sym - 7)*(sp.exp(2*x_sym)))
A = sp.limit(f, x_sym, 1)
print('lim = ' ,A)

x = np.linspace(-2,2,100)
f_np = sp.lambdify(x_sym, f, 'numpy')
y_vals = f_np(x)

plt.plot(x, y_vals, color='black')
plt.title('График функции (-x^2 - 3x - 7)*(e^2x)')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()
```

```python
# в)
f = ((sp.sin(x_sym))/(sp.sqrt(x_sym + 1) - sp.sqrt(2*x_sym + 1)))
A = sp.limit(f, x_sym, 0)
print('lim = ' ,A)

x = np.linspace(-2,2,100)
f_np = sp.lambdify(x_sym, f, 'numpy')
y_vals = f_np(x)

plt.plot(x, y_vals, color='black')
plt.title('График функции sin(x)/(sqrt(x + 1) - sqrt(2*x + 1))')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()
```

На самом деле при х стремится к бесконечности предел функции равен бесконечности. Правосторонний и левосторонний пределы также равны бесконечности. Поэтому в этой точке функция претерпевает разрыв второго рода.

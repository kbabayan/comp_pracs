Упражнение 1

```python
import sympy as sp
from sympy.abc import x

print('a)')
y = (2*x-1.0)/(x-1)
print('x->1+0')
sp.plot(y,(x,100,1))
print('x->1-0')
sp.plot(y,(x,-100,1))
print('x->1')
sp.plot((y,(x,-100,1)),(y,(x,10,1)))

print('x->2+0')
sp.plot(y,(x,100,2))
print('x->2-0')
sp.plot((y,(x,-10,1)),(y,(x,1,2)))
print('x->2')
sp.plot((y,(x,-100,1)),(y,(x,1,2)),(y,(x,100,2)))

print('b)')
y = (2*x-1)/sp.Abs(x-1)
print('x->1+0')
sp.plot(y,(x,100,1))
print('x->1-0')
sp.plot(y,(x,-100,1))
print('x->1')
sp.plot((y,(x,-100,1)),(y,(x,100,1)))

print('x->2+0')
sp.plot(y,(x,100,2))
print('x->2-0')
sp.plot((y,(x,-100,1)),(y,(x,1,2)))
print('x->2')
sp.plot((y,(x,-100,1)),(y,(x,1,2)),(y,(x,100,2)))
```

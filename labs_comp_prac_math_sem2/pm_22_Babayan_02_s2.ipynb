```python
import sympy as sp
from scipy import optimize
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,np.pi/2,100)
func = x*np.cos(x)
fig, ax = plt.subplots()
plt.plot(x, func)
plt.grid()
ax.set_title('f(x) = xcos(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
x = sp.symbols('x')
func = x*sp.cos(x)
S = sp.integrate(func, (x, 0, sp.pi/2))
sp.pprint(S)
```

```python
x = np.linspace(-4,1,100)
func1 = x**2 + 2*x
fig, ax = plt.subplots()
plt.plot(x, func1, label = 'func1')

func2 = 7 - 4*x - x**2
plt.plot(x, func2, label = 'func2')

coefficients = [2, 6, -7]
intersection_points_x = np.roots(coefficients)
intersection_points_y = intersection_points_x ** 2 + 2 * intersection_points_x
print(intersection_points_x, intersection_points_y)

plt.scatter(intersection_points_x, intersection_points_y, color='red', label='Intersection Points')
plt.grid()
ax.set_title('ex 2')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.show()
```

```python
x = sp.symbols('x')
func1 = x**2 + 2*x
func2 = 7 - 4*x - x**2

S = sp.integrate(func2 - func1, (x, -3.89791576, 0.89791576))
sp.pprint(S)
```

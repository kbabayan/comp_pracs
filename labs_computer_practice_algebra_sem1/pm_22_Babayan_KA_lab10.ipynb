Упражнение 3

```python
A = Matrix([[2,-2],[-2,5]])
T,D=A.diagonalize()

x1,y1=symbols('x1 y1')
X1=Matrix([[x1,y1]])
X= T/sqrt(5)*X1.T
print('x=',simplify(X[0,0]))
print('y=',simplify(X[1,0]))

F=2*X[0]**2-4*X[0]*X[1]+5*X[1]**2+8*X[0]-2*X[1]+9
F=simplify(F)
print(F)

var('x y')
plot_implicit(Eq(2*x**2-4*x*y+5*y**2+8*x-2*y,-9),title='В старой СК', xlabel='x',ylabel='y')
plot_implicit(Eq(0.5*x**2+3*y**2,1),title='В новой СК', xlabel='x1',ylabel='y1')

F
```

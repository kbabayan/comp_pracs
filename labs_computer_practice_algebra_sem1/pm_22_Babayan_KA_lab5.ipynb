Упражнение 5.4

```python
M=Point(1,-1,-3)
n=(2,-3,5)

P=Plane(M,normal_vector=n)
print('Уравнение плоскости:',P.equation())
```

```python
A=2;B=-3;C=5;D=10
f=lambda x,y:(-A*x-B*y-D)/C
fig=plt.figure(figsize=(10,10))
ax=fig.add_subplot(1,1,1,projection='3d')
xval=np.linspace(-4,4,100)
yval=np.linspace(-4,4,100)
x,y=np.meshgrid(xval,yval)

ax.quiver(*M, *n,color='Blue')

z=f(x,y)
surf=ax.plot_surface(x,y,z,rstride=1,cstride=1,cmap=cm.plasma)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Плоскость 2x - 3y + 5z + 10=0')
ax.view_init(10,-30)
plt.show()

p0=Point(0,0,0)
dist=P.distance(p0)
print('Расстояние от начала координат до плоскости:',dist)
```

```python
plane1_1=Plane(Point(11,0,0),normal_vector=(5,3,-9))

A=5;B=3;C=-9;D=-55
f=lambda x,y:(-A*x-B*y-D)/C
fig=plt.figure(figsize=(10,10))
ax=fig.add_subplot(1,1,1,projection='3d')
xval=np.linspace(-8,8,100)
yval=np.linspace(-8,8,100)
x,y=np.meshgrid(xval,yval)

p=np.array([11,0,0]) #точка на плоскости
v=np.array([5,3,-9]) #вектор нормали
ax.quiver(*p, *v, color='Blue') # строим вектор нормали

z = f(x, y) # вычисляем значения z в точках
surf = ax.plot_surface(x, y, z, rstride=1, cstride=1, cmap=cm.magma) # строим плоскость

plt.xlabel('x')
plt.ylabel('y')
plt.title('x - 10y - 6z + 4=0')
ax.view_init(200, -60)
plt.show()
print('Расстояние от точки N(-3,1,3) до плоскости:',plane1_1.distance(Point(-3,1,3)))
```

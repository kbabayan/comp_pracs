Упражнение 4.3

```python
x = np.linspace(-5, 5, 100) 

L1=Line((0,0),(2,3))
L1_eq=L1.equation()

P1=Point(-2,2)
L2=L1.parallel_line(P1)
L2_eq=L2.equation()

L3=L1.perpendicular_line(P1)
L3_eq=L3.equation()

print('L1: ',L1_eq)
print('L2: ',L2_eq)
print('L3: ',L3_eq)
plt.plot(x,3*x/2) #L1
plt.plot(x,(3*x+10)/2) #L2
plt.plot(x,(-2*x+2)/3) #L3
plt.xlabel('x')
plt.ylabel('y')
plt.grid ()
plt.axis('equal')
plt.legend(['L1', 'L2', 'L3'], loc = 2)
plt.show()
```

```python
import numpy as np
from sympy import *
import matplotlib.pyplot as plt
```

Упаражнение 1

```python
x = 2*np.sqrt(3)
y = -np.sqrt(3)
print(x,y)

a = -1
b = -2
alfa =np.pi/6
x1=(x-a)*cos(-alfa)-(y-b)*sin(-alfa)
y1=(x-a)*sin(-alfa)+(y-b)*cos(-alfa)
print(x1,y1)

x2 = x1 * cos(alfa) - y1 * sin(alfa)+a
y2 = x1 * sin(alfa) + y1 * cos(alfa)+b
print('Проверка:', x2, y2)
```

Упражнение 2

```python
a = 3; b = 2
x1, y1, alfa = symbols('x1 y1 alfa')
x = x1 * cos(alfa) + y1 * sin(alfa)
y = - y1 * cos(alfa) + x1 * sin(alfa)

f = 2 * x - 4 - y
y1 = solve(f, 'y1')

print('решение системы:', y1)
y1 = x1 * (-sin(alfa) + 2 * cos(alfa)) / (2 * sin(alfa) + cos(alfa))
yy = y1.subs(alfa, -pi / 4)
print('Уравнение повернутой прямой:', yy)

x = np.linspace(-1, 3, 100)
plt.figure(figsize=(10,10))
plt.plot(x+a, - sqrt(2) *(3* sqrt(2)*x/2 - 4)+b, c ='b')
plt.plot(x, 2 * x - 4, c = 'magenta')

alfa = np.pi/4
x0 = 1; y0 = -2
x2 = x0 * np.cos(alfa) - y0 * np.sin(alfa)+a
y2 = y0 * np.cos(alfa) + x0 * np.sin(alfa)+b
print('Координаты точки в новой системе:', x2, y2)

plt.arrow(-2, 0, 6, 0, width=0.05)
plt.arrow(0, -2, 0, 6, width=0.05)

x3 = np.cos(alfa); y3 = np.sin(alfa)
x4 = -np.sin(alfa); y4 = np.cos(alfa)

plt.arrow(0, 0, x3, y3, width=0.05)
plt.arrow(0, 0, x4, y4, width=0.05)

plt.plot(x0, y0, 'or')
plt.plot(x2, y2, 'or')
plt.axis('equal')
plt.text(x0, y0, 'M')
plt.text(x2, y2, 'M1')
plt.xlabel('X'); plt.ylabel('Y')
plt.grid()
```

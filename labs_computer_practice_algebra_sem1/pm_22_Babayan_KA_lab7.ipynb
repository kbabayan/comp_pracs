```python
import numpy as np
from sympy import *
import matplotlib.pyplot as plt
```

Упражнение 1. Эллипс

```python
e1 = Ellipse(Point(3, 1), hradius=3, eccentricity=Rational(4, 5))
a=e1.hradius
b=e1.vradius
E=e1.eccentricity
F=e1.foci
print('величина большей полуоси = ',a)
print('величина меньшей полуоси = ',b)
print('Эксцентриситет = ',E)
print('Координаты фокусов ',F)

t = np.arange(0, 2*np.pi, 0.01)
x = a*np.cos(t)+3
y = b*np.sin(t)+1
plt.plot(x, y, lw=2)
plt.ylabel('$y$')
plt.xlabel('$x$')
plt.plot(3/5,1,'o')
plt.plot(27/5,1,'o')
plt.plot(3,1,'o')
plt.grid()
```

Упражнение 2. Парабола

```python
plt.subplot(2,1,1)
y = np.arange(-20, 10, 0.01)
x1 = np.sqrt(-16*y)
x2 = -np.sqrt(-16*y)
plt.plot(x1, y+4, lw=2,color='c')
plt.plot(x2, y+4, lw=2,color='c')
plt.plot(0,0,'or')
x = np.linspace(-20,20,100)
plt.plot(x,x*0+8,'m')
plt.axis('equal')
plt.grid()
plt.show()

p = Parabola(Point(0, 0),Line(Point(0, 8), Point(1, 8)))
print('Фокус:',p.focus)
print('Фокальное расстояние =',p.focal_length)
print('Эксцентриситет =',p.eccentricity)
print('Директриса:',p.directrix)
print('Ось симметрии:',(p.axis_of_symmetry))
print('Каноническое уравнение',p.equation())
```

Упражнение 3. Гипербола

```python
a = np.sqrt(8)
b = np.sqrt(2)
c = np.sqrt(a**2+b**2)

x = np.linspace(-30,30,1000)
y = np.sqrt(2+((x)**2/4))
plt.plot(x+2, y, lw=2,color='r')
y = -np.sqrt(2+((x)**2/4))
plt.plot(x+2, y, lw=2,color='r')

y = b*(x)/a
plt.plot(x+2, y, '-b')
y = -b*(x)/a
plt.plot(x+2, y, '-b')
plt.plot(2,c,'or')
plt.plot(2,-c,'or')
plt.axis('equal')
plt.grid(True)
plt.show()
```
